```python
import pandas as pd
import sklearn
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris_data = load_iris()
iris_df = pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
iris_df['target'] = iris_data['target']
iris_df.head()
```

```python
iris_x = iris_df.iloc[:,:4].values
iris_y = iris_df.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_y = LabelEncoder()
iris_y = label_encoder_y.fit_transform(iris_y)
iris_y
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
iris_x = std_scaler.fit_transform(iris_x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
iris_x = pca.fit_transform(iris_x)
```

```python
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zdata = iris_x[:, 2]
xdata = iris_x[:, 0]
ydata = iris_x[:, 1]
ax.scatter3D(xdata, ydata, zdata, c=iris_y, cmap=plt.cm.coolwarm)
ax.set_title('Iris data in 3-dimension')
ax.set_xlabel("Projection on x-axis")
ax.set_ylabel("Projection on y-axis")
ax.set_zlabel("Projection on z-axis")
```
